```python
def lnprior(p):
    # The parameters are stored as a vector of values, so unpack them
    alpha,betax,eps = p
    # We're using only uniform priors, and only eps has a lower bound
    if eps <= 0:
        return -inf
    return 0

def lnlike(p, x, z, zerr):
    alpha,betax,eps = p
    model = alpha + betax*x
    # the likelihood is sum of the lot of normal distributions
    denom = np.power(zerr,2) + np.power(np.exp(eps),2)*np.power(model,2)
    lp = -0.5*sum(np.power((z - model),2)/denom + np.log(2*np.pi*denom))
    return lp

def lnprob(p, x, z):
    lp = lnprior(p)
    if not isfinite(lp):
        return -inf
    return lp + lnlike(p, x, z)
```

```python
import scipy.optimize as opt
nll = lambda *args: -lnlike(*args)
print(nll)
result = opt.minimize(nll, [b, m, np.mean(Table1[:,3])],
                      args=(Table1[:,1], Table1[:,2], Table1[:,3]))
print(result['x'])
```

```python
from numpy import *
Nwalker,Ndim = 50,3
p0 = [result['x']+1.e-4*random.randn(Ndim) for i in range(Nwalker)]
```

```python
import emcee
sampler = emcee.EnsembleSampler(Nwalker,Ndim,lnprob,
                                args=(Table1[:,1],Table1[:,2]))
pos,prob,state = sampler.run_mcmc(p0, 500)
```

```python
sampler.reset()
pos,prob,state = sampler.run_mcmc(pos, 1000)
```

```python
m_alpha,m_betax,m_eps = median(sampler.flatchain, axis=0)

plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.plot(Table1[:,1], Table1[:,2]-m_alpha, 'o')
plt.xlabel('X')
plt.ylabel('Z - alpha - beta_y y')
# Now plot the model
xx = array([Table1[:,1].min(), Table1[:,1].max()])
plt.plot(xx, xx*m_betax)
plt.plot(xx, xx*m_betax + m_eps, '--', color='k')
plt.plot(xx, xx*m_betax - m_eps, '--', color='k')
```
